```python
a=data.sort_values(by='Reviews',ascending=False)[['App','Category','Reviews']].head(10)
```

```python
a=a.drop('Category',axis=1)
```

```python
a['Reviews']=pd.to_numeric(a['Reviews'])
```

```python
a.plot(kind='bar',stacked=True,y='Reviews')
```
